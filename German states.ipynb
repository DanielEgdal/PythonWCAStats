```python
from collections import defaultdict
import pandas as pd
```

```python
comps = open("WCA_export_Competitions.tsv").readlines()[1:]
```

```python
for idx, persson in enumerate(comps):
    comps[idx] = persson.split('\t')
```

```python
# Keep all German comps and their listed city.
germany = []
for comp in comps:
    if comp[3] == 'Germany':
        germany.append([comp[0],comp[2]])
```

```python
# For mapping later, so the city can be tracked by comp ID
comp_cit = defaultdict()
for val in germany:
    comp_cit[val[0]] = val[1]
```

```python
# Storing all German comps in a set so the results table is easier to handle once all non-German comps are filtered out.
german_comps = set()
for val in germany:
    german_comps.add(val[0])
```

```python
# All distinct cities
city = set()
for comp in germany:
    city.add(comp[1])
```

```python
# Just a filler and initialization of the dictionary.
mapping = defaultdict()
for thing in city:
    mapping[thing] = 'State'
```

```python
# All states per person
persdic = defaultdict(set)
```

```python
results = pd.read_csv("WCA_export_Results.tsv", sep = '\t')
```

```python
results = results[results['competitionId'].isin(german_comps)]
```

```python
for val in results.values:
    persdic[val[7]].add(mapping[comp_cit[val[0]]])
```

```python
# Counting the amount of comps
score = defaultdict()
for val in persdic:
    score[val] = len(persdic[val])
```

```python
col1 = []
col2 = []
for val in score:
    col1.append(val)
    col2.append(score[val])
```

```python
data = {'ID': col1, 'GTS': col2}
GTS = pd.DataFrame.from_dict(data)
```
